## Задание 2
С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.

```python
t = np.linspace(0,10,51)
t
```

Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.

```python
f = np.cos(t)
f
```

Постройте линейную диаграмму, используя массив t для координат по горизонтали, а массив f - для координат по вертикали. Лния графика должна быть зеленого цвета.

```python
plt.plot(t,f)
plt.show()
```

Выведите название диаграммы - 'График f(t)'.

```python
plt.plot(t,f)
plt.title("График f(t)")
plt.show()
```


Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.

```python
plt.plot(t,f)
plt.title("График f(t)")
plt.xlabel("Значения t")
plt.ylabel("Значения f")
plt.show()
```

Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.

```python
plt.plot(t,f)
plt.title("График f(t)")
plt.xlabel("Значения t")
plt.ylabel("Значения f")
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

## * Задание 3
С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.

```python
x = np.linspace(-3,3,51)
```

Создайте массивы y1, y2, y3, y4 по следующим формулам:

    y1 = x**2

    y2 = 2 * x + 0.5

    y3 = -3 * x - 1.5
    
    y4 = sin(x)


```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig  и массив объектов Axes под названием ax,
причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов.

В каждом графике массив x используется для координат по горизонтали.

В левом верхнем графике для координат по вертикали используйте y1, в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.

Дайте название графикам: 'График y1', 'График y2' и т.д.

Для графика в левом верхнем углу установите границы по оси x от -5 до 5.

Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.

Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.


```python
fig, ax = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(8,6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x,y1)
ax1.set_title("График y1")
ax1.set_xlim([-5, 5])

ax2.plot(x,y2)
ax2.set_title("График y2")

ax3.plot(x,y3)
ax3.set_title("График y3")

ax4.plot(x,y4)
ax4.set_title("График y4")
```

```python
plt.style.use('fivethirtyeight')

import pandas as pd

kaggle = pd.read_csv('creditcard.csv')
kaggle.head()
```

```python
kaggle.describe()
```

```python
kaggle['Class'].value_counts()
```

```python
plt.hist(kaggle.loc[kaggle['Class']==0, 'V1'], bins = 20, density = True, alpha=0.5, color='grey')
plt.hist(kaggle.loc[kaggle['Class']==1, 'V1'], bins = 20, density = True, alpha=0.5, color='red')
plt.legend(['Class 0', 'Class 1'])
plt.xlabel('V1')
```
